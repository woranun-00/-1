## 1. ตัวแปร (Variables)

**ตัวแปร** คือ "ชื่อ" ที่เราใช้เก็บค่าข้อมูลไว้ในหน่วยความจำของคอมพิวเตอร์ เปรียบเหมือนกล่องที่มีป้ายชื่อติดไว้ — เราเก็บของ (ค่า) ไว้ในกล่อง แล้วเรียกของนั้นกลับมาใช้ผ่านชื่อกล่อง

ใช้เครื่องหมาย `=` (เรียกว่า assignment operator) ในการกำหนดค่าให้ตัวแปร:

```python

```

**ข้อสังเกต:** Python ไม่ต้องประกาศชนิดข้อมูลล่วงหน้า (ไม่เหมือน C/Java) — Python จะ "เดา" ชนิดข้อมูลให้เองจากค่าที่เราใส่


````python
# สร้างตัวแปรเก็บชื่อ อายุ และส่วนสูง
name = "สมชาย"
age = 20
height = 175.5

print(name)
print(age)
print(height)
````

### กฎการตั้งชื่อตัวแปร (Naming Rules)

**ข้อแนะนำ (Convention):** นิยมใช้ตัวพิมพ์เล็กและ `_` คั่นคำ เช่น `student_name` (เรียกว่า snake_case)


````python
# ตัวอย่างชื่อตัวแปรที่ดีและไม่ดี

# ดี ✅
student_name = "มานี"
total_score = 85

# ใช้ได้ แต่ไม่ใช่ convention ที่นิยม
StudentName = "มานี"   # นิยมใช้กับชื่อ class มากกว่า

print(student_name, total_score)
````

## 2. ชนิดข้อมูล (Data Types)

Python มีชนิดข้อมูลพื้นฐานหลายแบบ แต่ที่ใช้บ่อยที่สุด 4 แบบคือ:

| ชนิดข้อมูล | ชื่อเต็ม | ตัวอย่าง | ใช้เก็บอะไร |
|---|---|---|---|
| `int` | Integer | `10`, `-5`, `0` | เลขจำนวนเต็ม |
| `float` | Floating point | `3.14`, `-0.5` | เลขทศนิยม |
| `str` | String | `"สวัสดี"`, `'Hi'` | ข้อความ |
| `bool` | Boolean | `True`, `False` | ค่าจริง/เท็จ |

https://www.w3schools.com/python/python_datatypes.asp

ใช้ฟังก์ชัน `type()` เพื่อตรวจสอบชนิดข้อมูลของตัวแปรได้


````python
# ตรวจสอบชนิดข้อมูลด้วย type()
a = 10
b = 3.14
c = "Data Science"
d = True

print(type(a))   # int
print(type(b))   # float
print(type(c))   # str
print(type(d))   # bool
````

### 2.1 int (จำนวนเต็ม)

ใช้เก็บเลขจำนวนเต็ม ทั้งบวกและลบ ไม่มีจุดทศนิยม


````python
# ตัวอย่าง int
num_students = 45
temperature_celsius = -2
year = 2569

print(num_students, temperature_celsius, year)
print(type(num_students))
````

### 2.2 float (ทศนิยม)

ใช้เก็บเลขที่มีจุดทศนิยม — แม้จะใส่ `.0` ก็ถือเป็น float ทันที


````python
# ตัวอย่าง float
gpa = 3.75
pi_value = 3.14159
price = 99.0   # มี .0 ก็เป็น float แล้ว

print(gpa, pi_value, price)
print(type(price))
````

### 2.3 str (ข้อความ)

ใช้เก็บข้อความ ครอบด้วยเครื่องหมายคำพูดเดี่ยว `'...'` หรือคู่ `"..."` ก็ได้ — ผลลัพธ์เหมือนกัน


````python
# ตัวอย่าง str
greeting = "สวัสดีค่ะ"
course_code = 'SC 612 104'

print(greeting)
print(course_code)

# เชื่อมข้อความด้วย +
full_text = greeting + " " + course_code
print(full_text)

# str ความยาวหลายบรรทัด ใช้ '''...''' หรือ """..."""
message = """บรรทัดที่ 1
บรรทัดที่ 2"""
print(message)
````

### 2.4 bool (ค่าจริง/เท็จ)

มีได้แค่ 2 ค่า: `True` หรือ `False` (ตัวอักษรตัวแรกต้องพิมพ์ใหญ่เสมอ) มักเกิดจากการเปรียบเทียบค่า


````python
# ตัวอย่าง bool
is_student = True
is_raining = False

print(is_student, is_raining)
print(type(is_student))

# bool มักเกิดจากการเปรียบเทียบ
score = 75
is_passed = score >= 50
print("ผ่านไหม:", is_passed)
````

เวลาต้องการแสดงข้อความผสมกับค่าตัวแปร ใช้ `f"..."` แล้วใส่ตัวแปรใน `{}` จะอ่านง่ายกว่าการ `+` ข้อความเข้าด้วยกันมาก


````python
name = "มานี"
age = 20
gpa = 3.75

# วิธีเก่า (เชื่อมด้วย +) — ต้องแปลงชนิดข้อมูลเองและอ่านยาก
print("ชื่อ " + name + " อายุ " + str(age) + " ปี")

# วิธีใหม่ (f-string) — อ่านง่าย ไม่ต้องแปลงชนิดข้อมูลเอง
print(f"ชื่อ {name} อายุ {age} ปี เกรดเฉลี่ย {gpa}")
````

## 3. Operators (ตัวดำเนินการ)


แบ่งเป็น 3 กลุ่มหลักที่ต้องรู้จัก: **คณิตศาสตร์**, **เปรียบเทียบ**, และ **ตรรกะ**

### 3.1 Arithmetic Operators (ตัวดำเนินการทางคณิตศาสตร์)

| Operator | ความหมาย | ตัวอย่าง | ผลลัพธ์ |
|---|---|---|---|
| `+` | บวก | `5 + 3` | `8` |
| `-` | ลบ | `5 - 3` | `2` |
| `*` | คูณ | `5 * 3` | `15` |
| `/` | หาร (ได้ float เสมอ) | `5 / 2` | `2.5` |
| `//` | หารปัดเศษทิ้ง (floor division) | `5 // 2` | `2` |
| `%` | หารเอาเศษ (modulo) | `5 % 2` | `1` |
| `**` | ยกกำลัง | `5 ** 2` | `25` |

````python
a = 17
b = 5

print("a + b  =", a + b)
print("a - b  =", a - b)
print("a * b  =", a * b)
print("a / b  =", a / b)    # หารปกติ ได้ float
print("a // b =", a // b)   # หารปัดเศษทิ้ง
print("a % b  =", a % b)    # เศษจากการหาร
print("a ** b =", a ** b)   # a ยกกำลัง b
````

**จุดที่ต้องระวัง:** `/` กับ `//` ให้ผลต่างกัน!
- `/` คือหารแบบปกติ ผลลัพธ์เป็น `float` เสมอ แม้จะหารลงตัว
- `//` คือหารแบบปัดเศษทิ้ง (floor) ผลลัพธ์เป็น `int` ถ้า input เป็น int ทั้งคู่

`%` (modulo) มีประโยชน์มากในการเช็คว่าเลขคู่/เลขคี่ หรือเช็คตัวหารร่วม


````python
# ตัวอย่างการใช้ % เช็คเลขคู่/เลขคี่
number = 17

if number % 2 == 0:
    print(f"{number} เป็นเลขคู่")
else:
    print(f"{number} เป็นเลขคี่")
````

#### ลำดับการคำนวณ (Order of Operations / PEMDAS)

เมื่อมีตัวดำเนินการหลายตัวอยู่ในนิพจน์เดียว Python จะคำนวณตามลำดับความสำคัญ (Precedence)
เหมือนที่เราคุ้นเคยในคณิตศาสตร์ ซึ่งเรียกย่อว่า PEMDAS/BODMAS

- **P**arentheses (วงเล็บ `()`) — ทำในวงเล็บก่อนเสมอ
- **E**xponents (เลขยกกำลัง `**`) — ทำเป็นลำดับถัดมา
- **M**ultiplication (คูณ `*`) และ **D**ivision (หาร `/`, `//`, `%`) — มีความสำคัญเท่ากัน ทำจากซ้ายไปขวา
- **A**ddition (บวก `+`) และ **S**ubtraction (ลบ `-`) — มีความสำคัญเท่ากัน ทำจากซ้ายไปขวา

การใช้วงเล็บ `()` สามารถบังคับลำดับการคำนวณให้เป็นไปตามที่เราต้องการได้

--- ตัวอย่าง PEMDAS ---


````python
# ไม่มีวงเล็บ: คูณ/หาร ทำก่อน บวก/ลบ
# 10 + 5 * 2 => 10 + 10 => 20
result1 = 10 + 5 * 2
print(f"10 + 5 * 2 = {result1}")
````

````python
# มีวงเล็บ: ทำในวงเล็บก่อน
# (10 + 5) * 2 => 15 * 2 => 30
result2 = (10 + 5) * 2
print(f"(10 + 5) * 2 = {result2}")
````

````python
# ตัวอย่างที่ซับซ้อนขึ้น:
result3 = 2 ** 3 + 6 / 3 - 1
print(f"2 ** 3 + 6 / 3 - 1 = {result3}")
````

````python
# ใช้วงเล็บเพื่อเปลี่ยนลำดับ
result4 = 2 ** (3 + 6) / 3 - 1
print(f"2 ** (3 + 6) / 3 - 1 = {result4}")
````

````python
print("\nข้อควรจำ: เมื่อมี operator ความสำคัญเท่ากัน ให้ทำจากซ้ายไปขวา")
# 10 / 2 * 5
result5 = 10 / 2 * 5
print(f"10 / 2 * 5 = {result5}")
````

````python
# 10 * 5 / 2
result6 = 10 * 5 / 2
print(f"10 * 5 / 2 = {result6}")
````

### 3.2 Comparison Operators (ตัวดำเนินการเปรียบเทียบ)

ผลลัพธ์จากการเปรียบเทียบเป็น `bool` เสมอ (`True` หรือ `False`)

| Operator | ความหมาย | ตัวอย่าง | ผลลัพธ์ |
|---|---|---|---|
| `==` | เท่ากับ | `5 == 5` | `True` |
| `!=` | ไม่เท่ากับ | `5 != 3` | `True` |
| `>` | มากกว่า | `5 > 3` | `True` |
| `<` | น้อยกว่า | `5 < 3` | `False` |
| `>=` | มากกว่าหรือเท่ากับ | `5 >= 5` | `True` |
| `<=` | น้อยกว่าหรือเท่ากับ | `5 <= 3` | `False` |

**⚠️ ข้อผิดพลาดที่พบบ่อย:** `=` ใช้กำหนดค่า ส่วน `==` ใช้เปรียบเทียบ — สลับกันใช้ไม่ได้!


````python
x = 10
y = 20

print(x == y)   # False เพราะ 10 ไม่เท่ากับ 20
print(x != y)   # True เพราะ 10 ไม่เท่ากับ 20 จริง
print(x < y)    # True
print(x >= 10)  # True เพราะ 10 เท่ากับ 10
````

### 3.3 Logical Operators (ตัวดำเนินการตรรกะ)

ใช้รวมเงื่อนไขหลายๆ อันเข้าด้วยกัน

| Operator | ความหมาย | ตัวอย่าง |
|---|---|---|
| `and` | จริงทั้งคู่ จึงเป็นจริง | `True and False` → `False` |
| `or` | จริงอย่างน้อยหนึ่ง จึงเป็นจริง | `True or False` → `True` |
| `not` | กลับค่าตรงข้าม | `not True` → `False` |


````python
age = 20
has_id_card = True

# ต้องอายุ >= 18 AND มีบัตรประชาชน จึงเข้าได้
can_enter = (age >= 18) and (has_id_card == True)
print("เข้าได้ไหม:", can_enter)

# ตัวอย่าง or
is_weekend = False
is_holiday = True
can_rest = is_weekend or is_holiday
print("วันนี้พักได้ไหม:", can_rest)

# ตัวอย่าง not
is_raining = False
print("ไม่ฝนตกใช่ไหม:", not is_raining)
````

## 4. Type Conversion / Casting(การแปลงชนิดข้อมูล)

บางครั้งข้อมูลที่ได้มาไม่ใช่ชนิดที่เราต้องการ เช่น รับค่าจากผู้ใช้มาเป็น `str` เสมอ แต่เราต้องเอาไปคำนวณเป็นตัวเลข — ต้องแปลงชนิดข้อมูลก่อน

| ฟังก์ชัน | แปลงเป็น | ตัวอย่าง |
|---|---|---|
| `int()` | จำนวนเต็ม | `int("5")` → `5` |
| `float()` | ทศนิยม | `float("3.14")` → `3.14` |
| `str()` | ข้อความ | `str(5)` → `"5"` |
| `bool()` | จริง/เท็จ | `bool(1)` → `True` |


````python
# str -> int / float
age_text = "20"y
age_number = int(age_text)
print(age_number + 5)   # บวกได้แล้วเพราะเป็น int

price_texst = "99.50"
price_number = float(price_text)
print(price_number * 2)

# int/float -> str (ใช้บ่อยตอน + ข้อความ)
score = 95
message = "คะแนนของคุณคือ " + str(score)
print(message)
````

**⚠️ ข้อผิดพลาดที่พบบ่อย:** ถ้า string ไม่ใช่ตัวเลข จะแปลงไม่ได้ และ Python จะแจ้ง error แบบ `ValueError`


````python
# ทดลองรันเซลล์นี้ดู จะเกิด error อะไร?
text = "สวัสดี"
# number = int(text)   # ลบ # ข้างหน้าแล้วรันดู จะเห็น ValueError

# วิธีป้องกัน: เช็คก่อนแปลงด้วย .isdigit()
text2 = "123"
if text2.isdigit():
    print(int(text2) + 1)
else:
    print("แปลงเป็นตัวเลขไม่ได้")
````

````python
# bool <-> int: True เท่ากับ 1, False เท่ากับ 0
print(int(True))    # 1
print(int(False))   # 0
print(bool(0))       # False
print(bool(1))       # True
print(bool(99))      # True -- ตัวเลขที่ไม่ใช่ 0 ทั้งหมดเป็น True!
print(bool(""))      # False -- string ว่างเป็น False
print(bool("hello")) # True -- string ที่ไม่ว่างเป็น True
````

## 5. if / elif / else (การเขียนเงื่อนไข)

ใช้ตัดสินใจว่าโปรแกรมควรทำอะไร ขึ้นอยู่กับเงื่อนไขที่กำหนด

### โครงสร้างพื้นฐาน

```python
if เงื่อนไข1:
    # ทำสิ่งนี้ ถ้าเงื่อนไข1 เป็นจริง
elif เงื่อนไข2:
    # ทำสิ่งนี้ ถ้าเงื่อนไข1 ไม่จริง แต่เงื่อนไข2 เป็นจริง
else:
    # ทำสิ่งนี้ ถ้าไม่มีเงื่อนไขไหนเป็นจริงเลย
```

**⚠️ จุดสำคัญที่สุดของ Python:** ต้องมี **เครื่องหมาย `:`** ท้ายบรรทัดเงื่อนไข และ**บรรทัดถัดไปต้องเยื้อง (indent)** ด้วย Tab หรือเว้นวรรค 4 ครั้ง — Python ใช้การเยื้องบรรทัดแทนวงเล็บปีกกา `{}` แบบภาษาอื่น


````python
# if แบบพื้นฐาน
temperature = 35

if temperature > 30:
    print("วันนี้อากาศร้อนมาก")
````

````python
# if-else
score = 45

if score >= 50:
    print("ผ่าน")
else:
    print("ไม่ผ่าน")
````

````python
# if-elif-else: ใช้ตัดเกรดจากคะแนน
score = 78

if score >= 80:
    grade = "A"
elif score >= 70:
    grade = "B"
elif score >= 60:
    grade = "C"
elif score >= 50:
    grade = "D"
else:
    grade = "F"

print(f"คะแนน {score} ได้เกรด {grade}")
````

### เงื่อนไขซ้อนกัน (Nested if) และเงื่อนไขซับซ้อน

สามารถใส่ `if` ซ้อนใน `if` ได้ และรวมหลายเงื่อนไขด้วย `and` / `or`


````python
# nested if: เช็คสมัครสมาชิกแบบมีเงื่อนไขซ้อน
age = 16
has_parent_consent = True

if age >= 18:
    print("สมัครได้เลย")
else:
    if has_parent_consent:
        print("อายุไม่ถึง แต่มีผู้ปกครองยินยอม สมัครได้")
    else:
        print("สมัครไม่ได้ ต้องรอผู้ปกครองยินยอมก่อน")
````

````python
# รวมเงื่อนไขด้วย and / or ในบรรทัดเดียว (อ่านง่ายกว่า nested if)
age = 16
has_parent_consent = True

if age >= 18 or (age >= 13 and has_parent_consent):
    print("สมัครได้")
else:
    print("สมัครไม่ได้")
````
