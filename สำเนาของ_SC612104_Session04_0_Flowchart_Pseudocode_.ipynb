## 1. Flowchart และ Pseudocode คืออะไร?

ก่อนเขียนโค้ดจริง โปรแกรมเมอร์มักจะ "ออกแบบความคิด" ก่อนด้วย 2 เครื่องมือนี้:

| | Flowchart | Pseudocode |
|---|---|---|
| **คืออะไร** | แผนภาพแสดงลำดับการทำงานด้วยรูปทรงและลูกศร | คำบรรยายขั้นตอนการทำงาน เขียนคล้ายโค้ดแต่ใช้ภาษาธรรมดา |
| **เหมาะกับ** | คนที่ถนัดดูภาพ เห็นลำดับและทิศทางชัดเจน | คนที่ถนัดอ่านเป็นขั้นตอน ใกล้เคียงโค้ดจริงมากกว่า |
| **ใช้ตอนไหน** | ออกแบบ logic ที่ซับซ้อน มีหลายทางแยก | วางแผนลำดับขั้นตอนก่อนแปลงเป็นโค้ด |

**ทำไมต้องรู้ทั้งสองแบบ?**
เพราะการ "คิดก่อนเขียนโค้ด" ช่วยลดข้อผิดพลาด และทำให้ debug ง่ายขึ้น — เวลาโค้ดมีปัญหา เราย้อนไปดู Flowchart/Pseudocode ได้ว่า logic ตรงไหนผิดไปจากที่วางแผนไว้

> 🎯 **เป้าหมายของ Notebook นี้:** ฝึกอ่านและเขียน Flowchart/Pseudocode ให้คล่อง แล้วฝึกแปลงเป็นโค้ด Python จริงให้ได้

## 2. สัญลักษณ์มาตรฐานของ Flowchart

| สัญลักษณ์ | รูปทรง | ความหมาย |
|---|---|---|
| 🟢 Start/End | วงรี (Oval) | จุดเริ่มต้น/จุดสิ้นสุดของโปรแกรม |
| 🟦 Process | สี่เหลี่ยมผืนผ้า (Rectangle) | ขั้นตอนการประมวลผล/คำนวณ/กำหนดค่า |
| 🔶 Decision | สี่เหลี่ยมข้าวหลามตัด (Diamond) | จุดตัดสินใจ (ตอบ Yes/No หรือ True/False) |
| ⬜ Input/Output | สี่เหลี่ยมด้านขนาน (Parallelogram) | รับข้อมูลเข้า/แสดงผลลัพธ์ |
| ➡️ Flow line | ลูกศร (Arrow) | ทิศทางการไหลของโปรแกรม |

ใน Notebook นี้ใช้ภาษา **Mermaid** ในการวาด ซึ่งจะใช้สัญลักษณ์แบบง่าย:
- `([...])` = Start/End (วงรี)
- `[...]` = Process (สี่เหลี่ยม)
- `{...}` = Decision (ข้าวหลามตัด)
- `-->` = ลูกศรแสดงทิศทาง

## 3. if / elif / else

### 3.1 Flowchart พื้นฐาน: if-else เดี่ยว

โครงสร้างที่ง่ายที่สุด: เช็คเงื่อนไขเดียว แล้วแยกเป็น 2 ทาง (จริง/เท็จ)




<details><summary>📄 ดู Mermaid source code (กดเพื่อขยาย — เอาไปแก้ไขเองได้)</summary>

```mermaid
flowchart TD
    A([Start]) --> B[/รับค่า score/]
    B --> C{score >= 50?}
    C -->|True| D[/print: "ผ่าน"/]
    C -->|False| E[/print: "ไม่ผ่าน"/]
    D --> F([End])
    E --> F
```

</details>


### Pseudocode ของ if-else เดี่ยว

```
START
    INPUT score
    IF score >= 50 THEN
        OUTPUT "ผ่าน"
    ELSE
        OUTPUT "ไม่ผ่าน"
    END IF
END
```

### แปลงเป็นโค้ด Python จริง (จาก Session 2)


````python
score = 45

if score >= 50:
    print("ผ่าน")
else:
    print("ไม่ผ่าน")
````

**สังเกตความเชื่อมโยง:**
- รูปข้าวหลามตัด `{score >= 50?}` → ตรงกับ `if score >= 50:` ในโค้ด
- เส้นทาง `True`/`False` → ตรงกับ `if` / `else` block
- กล่อง Process `[/print: ...`/]` → ตรงกับคำสั่ง `print()`

---

### 3.2 Flowchart: if-elif-else หลายทาง (ตัดเกรด)

เมื่อมีมากกว่า 2 ทางเลือก ใช้ `elif` ต่อกันเป็นลำดับ — ใน Flowchart จะเห็นเป็นข้าวหลามตัดต่อกันเป็นทอดๆ




<details><summary>📄 ดู Mermaid source code (กดเพื่อขยาย — เอาไปแก้ไขเองได้)</summary>

```mermaid
flowchart TD
    A([Start]) --> B[/รับค่า score/]
    B --> C{score >= 80?}
    C -->|True| D[/grade = "A"/]
    C -->|False| E{score >= 70?}
    E -->|True| F[/grade = "B"/]
    E -->|False| G{score >= 60?}
    G -->|True| H[/grade = "C"/]
    G -->|False| I{score >= 50?}
    I -->|True| J[/grade = "D"/]
    I -->|False| K[/grade = "F"/]
    D --> L[/print grade/]
    F --> L
    H --> L
    J --> L
    K --> L
    L --> M([End])
```

</details>


### Pseudocode ของ if-elif-else (ตัดเกรด)

```
START
    INPUT score
    IF score >= 80 THEN
        grade = "A"
    ELSE IF score >= 70 THEN
        grade = "B"
    ELSE IF score >= 60 THEN
        grade = "C"
    ELSE IF score >= 50 THEN
        grade = "D"
    ELSE
        grade = "F"
    END IF
    OUTPUT grade
END
```

### แปลงเป็นโค้ด Python

**จุดสำคัญ:** สังเกตว่า Flowchart "ไหลลงเป็นทอดๆ" ทางขวาทุกครั้งที่เงื่อนไขเป็น False — นี่คือสิ่งที่ `elif` ทำ: เช็คเงื่อนไขถัดไปก็ต่อเมื่อเงื่อนไขก่อนหน้าเป็น False เท่านั้น


````python
0score = 78

if score >= 80:
    grade = "A"
elif score >= 70:
    grade = "B"
elif score >= 60:
    grade = "C"
elif score >= 50:
    grade = "D"
else:
    grade = "F"

print(f"คะแนน {score} ได้เกรด {grade}")
````

> 💡 **ลองสังเกต:** ถ้าเขียน Flowchart ผิดเป็นข้าวหลามตัดที่**แยกอิสระจากกัน** (ไม่ได้ไหลต่อกันเป็นทอด) จะเทียบเท่ากับเขียนโค้ดเป็น `if` หลายอันแยกกัน (ไม่ใช่ `elif`) ซึ่งจะทำงาน**ต่างจาก if-elif-else** เพราะทุกเงื่อนไขจะถูกเช็คทั้งหมดไม่ว่าอันก่อนหน้าจะจริงหรือไม่ — นี่เป็นจุดที่นักศึกษามักสับสนเวลาแปลง Flowchart เป็นโค้ด ต้องระมัดระวัง

## 4. for Loop

### Flowchart ของ for loop

`for` loop คือ loop ที่รู้จำนวนรอบแน่นอนล่วงหน้า (ผ่าน `range()` หรือความยาวของ list/string) — Flowchart จะเน้นที่ "ตัวนับ" (counter) วิ่งจากค่าเริ่มต้นไปจนครบ




<details><summary>📄 ดู Mermaid source code (กดเพื่อขยาย — เอาไปแก้ไขเองได้)</summary>

```mermaid
flowchart TD
    A([Start]) --> B[/total = 0/]
    B --> C[/num = 1/]
    C --> D{num <= 10?}
    D -->|True| E[/total = total + num/]
    E --> F[/num = num + 1/]
    F --> D
    D -->|False| G[/print total/]
    G --> H([End])
```

</details>


### Pseudocode ของ for loop (หาผลรวม 1 ถึง 10)

```
START
    total = 0
    FOR num = 1 TO 10
        total = total + num
    END FOR
    OUTPUT total
END
```

### แปลงเป็นโค้ด Python

**จุดสำคัญที่ต้องสังเกต:** ใน Python เราไม่ต้องเขียนกล่อง "ตั้งค่าตัวนับ" และ "เพิ่มค่าตัวนับ" เอง — `range()` กับ `for` จัดการให้อัตโนมัติ! Flowchart ด้านบนแสดง "สิ่งที่เกิดขึ้นจริงเบื้องหลัง" แม้โค้ด Python จะดูสั้นกว่ามาก


````python
total = 0

for num in range(1, 11):
    total = total + num

print(f"ผลรวม 1 ถึง 10 คือ {total}")
````

### Flowchart: for loop ที่มี if ซ้อนอยู่ข้างใน (เช็คเลขคู่/คี่)

Loop กับ Condition มักใช้งานร่วมกันเสมอ — ตัวอย่างนี้แสดง `for` ที่มี `if` ซ้อนอยู่ข้างใน เพื่อกรองว่าจะ print เฉพาะเลขคี่




<details><summary>📄 ดู Mermaid source code (กดเพื่อขยาย — เอาไปแก้ไขเองได้)</summary>

```mermaid
flowchart TD
    A([Start]) --> B[/num = 1/]
    B --> C{num <= 10?}
    C -->|True| D{num is even?}
    D -->|True| E[skip - continue]
    D -->|False| F[/print num/]
    E --> G[/num = num + 1/]
    F --> G
    G --> C
    C -->|False| H([End])
```

</details>


### Pseudocode

```
START
    FOR num = 1 TO 10
        IF num is even THEN
            CONTINUE  (ข้ามไปรอบถัดไป)
        ELSE
            OUTPUT num
        END IF
    END FOR
END
```

### แปลงเป็นโค้ด Python จริง (จาก Session 3)


````python
for num in range(1, 11):
    if num % 2 == 0:
        continue
    print(num)
````

## 5. while Loop

### Flowchart ของ while loop

`while` loop คือ loop ที่ **ไม่รู้จำนวนรอบล่วงหน้า** แต่รู้ "เงื่อนไข" ที่ใช้ตัดสินว่าจะทำงานต่อไปหรือหยุด — Flowchart จะวนกลับไปเช็คเงื่อนไขซ้ำๆ จนกว่าจะเป็น False




<details><summary>📄 ดู Mermaid source code (กดเพื่อขยาย — เอาไปแก้ไขเองได้)</summary>

```mermaid
flowchart TD
    A([Start]) --> B[/count = 1/]
    B --> C{count <= 5?}
    C -->|True| D[/print count/]
    D --> E[/count = count + 1/]
    E --> C
    C -->|False| F([End])
```

</details>


### Pseudocode ของ while loop

```
START
    count = 1
    WHILE count <= 5 DO
        OUTPUT count
        count = count + 1
    END WHILE
END
```

### แปลงเป็นโค้ด Python

**⚠️ จุดที่ Flowchart ช่วยป้องกัน infinite loop ได้ดีมาก:** สังเกตว่าในแผนภาพ มีลูกศรจาก "count = count + 1" วนกลับไปเช็คเงื่อนไขเสมอ — ถ้าวาด Flowchart แล้วลืมวาดกล่อง "เพิ่มค่า count" จะเห็นภาพชัดเจนว่าลูกศรวนกลับไปที่เงื่อนไขแบบไม่มีอะไรเปลี่ยน นั่นคือสัญญาณของ **infinite loop**


````python
count = 1

while count <= 5:
    print(f"count = {count}")
    count += 1
````

### Flowchart: while loop กับสถานการณ์ "รอจนกว่าจะถูก" (password)

ตัวอย่างจาก Session 3 — ไม่รู้จำนวนรอบล่วงหน้า (ไม่รู้ว่าผู้ใช้จะพิมพ์ถูกตอนไหน) จึงเหมาะกับ `while` มากกว่า `for`




<details><summary>📄 ดู Mermaid source code (กดเพื่อขยาย — เอาไปแก้ไขเองได้)</summary>

```mermaid
flowchart TD
    A([Start]) --> B[/guess = ตัวที่ 0/]
    B --> C{guess != password?}
    C -->|True| D[/print: รหัสผ่านผิด/]
    D --> E[/i = i + 1, guess = ตัวถัดไป/]
    E --> C
    C -->|False| F[/print: เข้าสู่ระบบสำเร็จ/]
    F --> G([End])
```

</details>


### Pseudocode

```
START
    correct_password = "python2025"
    i = 0
    guess = attempts[i]
    WHILE guess != correct_password DO
        OUTPUT "รหัสผ่านผิด: " + guess
        i = i + 1
        guess = attempts[i]
    END WHILE
    OUTPUT "เข้าสู่ระบบสำเร็จ"
END
```

### แปลงเป็นโค้ด Python จริง (จาก Session 3)


````python
correct_password = "python2025"
attempts = ["12345", "admin", "python2025"]
i = 0

guess = attempts[i]
while guess != correct_password:
    print(f"รหัสผ่านผิด: {guess}")
    i += 1
    guess = attempts[i]

print(f"รหัสผ่านถูกต้อง: {guess} -- เข้าสู่ระบบสำเร็จ!")
````

## 6. เปรียบเทียบ for vs while แบบ Flowchart คู่กัน

ทั้งสองทำงาน**เหมือนกัน** (print เลข 1-5) แต่ใช้ logic วาด Flowchart ต่างกันเล็กน้อย — สังเกตว่า `for` ซ่อนกล่อง "ตั้งค่าเริ่มต้น" และ "เพิ่มค่าตัวนับ" ไว้ในตัว ส่วน `while` ต้องเขียนทั้งสองกล่องนี้ออกมาให้เห็นชัดเจน




<details><summary>📄 ดู Mermaid source code (กดเพื่อขยาย — เอาไปแก้ไขเองได้)</summary>

```mermaid
flowchart LR
    subgraph FOR["for loop"]
    direction TB
    A1([Start]) --> B1[/i in range 1 to 6/]
    B1 --> C1{มีค่าถัดไปไหม?}
    C1 -->|True| D1[/print i/]
    D1 --> B1
    C1 -->|False| E1([End])
    end

    subgraph WHILE["while loop"]
    direction TB
    A2([Start]) --> B2[/i = 1/]
    B2 --> C2{i <= 5?}
    C2 -->|True| D2[/print i/]
    D2 --> E2[/i = i + 1/]
    E2 --> C2
    C2 -->|False| F2([End])
    end
```

</details>


### Pseudocode เทียบกัน

```
# for loop                          # while loop
START                                START
    FOR i = 1 TO 5                      i = 1
        OUTPUT i                        WHILE i <= 5 DO
    END FOR                                 OUTPUT i
END                                          i = i + 1
                                         END WHILE
                                     END
```

### โค้ด Python เทียบกัน


````python
# for loop
print("for loop:")
for i in range(1, 6):
    print(i)

print()

# while loop (ทำงานแบบเดียวกัน)
print("while loop:")
i = 1
while i <= 5:
    print(i)
    i += 1
````

**สรุปการเลือกใช้:**

| ใช้ `for` เมื่อ... | ใช้ `while` เมื่อ... |
|---|---|
| รู้จำนวนรอบแน่นอน | ไม่รู้จำนวนรอบ รู้แต่เงื่อนไขที่จะหยุด |
| วนผ่านสมาชิกใน list/string/range | รอ input จนกว่าจะถูกต้อง |
| ทำงานกับข้อมูลที่มีอยู่แล้ว | ทำงานจนกว่าจะเกิดเหตุการณ์บางอย่าง |

## 🧪 แบบฝึกหัด: วาด Flowchart และเขียน Pseudocode เอง

> ลองวาด Flowchart ด้วยมือ (หรือเขียนเป็น Mermaid code ก็ได้ถ้าอยากฝึกเพิ่ม) แล้วเขียน Pseudocode ก่อนแปลงเป็นโค้ด Python ของตัวเอง ลองรันโค้ดของตัวเองในเซลล์ที่เตรียมไว้เพื่อตรวจสอบว่าตรงกับ Flowchart ที่วาดไว้ไหม

### ข้อ 1: if/elif/else — ระดับ BMI
จากสูตร BMI ในแบบฝึกหัด Session 2 (`BMI = น้ำหนัก / ส่วนสูง²`) ลองวาด **Flowchart** และเขียน **Pseudocode** สำหรับการแปลผล BMI เป็น 4 ระดับ:
- BMI < 18.5 → "น้ำหนักน้อย"
- 18.5 ≤ BMI < 23 → "น้ำหนักปกติ"
- 23 ≤ BMI < 25 → "น้ำหนักเกิน"
- BMI ≥ 25 → "อ้วน"

จากนั้นเขียนโค้ด Python ให้ตรงกับ Flowchart ที่วาด


````python
# เขียนคำตอบข้อ 1 (โค้ด Python) ที่นี่

weight = 65
height = 1.70

bmi = weight / (height ** 2)

if bmi < 18.5:
    print("น้ำหนักน้อย")
elif bmi < 23:
    print("น้ำหนักปกติ")
elif bmi < 25:
    print("น้ำหนักเกิน")
else:
    print("อ้วน")
````

### ข้อ 2: for loop — ตัวคูณร่วมน้อย (เลขคูณ 3 หรือ 5)
วาด **Flowchart** และเขียน **Pseudocode** สำหรับโปรแกรมที่หาผลรวมของเลขทั้งหมดตั้งแต่ 1 ถึง 50 ที่หารด้วย 3 **หรือ** 5 ลงตัว

จากนั้นเขียนโค้ด Python ให้ตรงกับ Flowchart ที่วาด


````python
# เขียนคำตอบข้อ 2 (โค้ด Python) ที่นี่
total = 0

for i in range(1, 51):
    if i % 3 == 0 or i % 5 == 0:
        total += i

print(total)
````

### ข้อ 3: while loop — ลดเงินต้นด้วยดอกเบี้ย
สมมติมีเงินฝาก 1,000 บาท ได้ดอกเบี้ย 5% ต่อปี (ทบต้น) วาด **Flowchart** และเขียน **Pseudocode** สำหรับโปรแกรมที่หาว่า "ต้องฝากกี่ปี เงินจึงจะเกิน 2,000 บาท"

💡 Hint: ใช้ `while` เพราะไม่รู้จำนวนปีล่วงหน้า สูตรคำนวณดอกเบี้ยทบต้น: `เงินใหม่ = เงินเดิม * 1.05`

จากนั้นเขียนโค้ด Python ให้ตรงกับ Flowchart ที่วาด


````python
# เขียนคำตอบข้อ 3 (โค้ด Python) ที่นี่
money = 1000
years = 0

while money <= 2000:
    money = money * 1.05
    years += 1

print("ต้องฝาก", years, "ปี")
print("เงิน =", round(money, 2), "บาท")

````

### ข้อ 4: ท้าทายเพิ่มเติม (Challenge) — for ซ้อน if ซ้อน while
วาด **Flowchart** สำหรับโปรแกรมเกม "ทอยลูกเต๊า 3 ครั้ง" โดยมีเงื่อนไข:
- ทอยทีละครั้ง (จำลองด้วยเลขที่กำหนดไว้ล่วงหน้า เช่น `[3, 6, 2]`)
- ถ้าทอยได้ 6 → print "ยินดีด้วย ได้เลข 6!" และ**หยุดทอยทันที** (ใช้ `break`)
- ถ้าทอยได้เลขอื่น → print เลขที่ทอยได้ แล้วทอยต่อ
- ถ้าทอยครบ 3 ครั้งแล้วไม่ได้ 6 เลย → print "หมดโอกาสทอยแล้ว"

โจทย์นี้รวม loop + condition + break เข้าด้วยกัน — เป็นแบบฝึกหัดที่ดีในการฝึกอ่าน logic ที่ซับซ้อนขึ้น


````python
# เขียนคำตอบข้อ 4 (โค้ด Python) ที่นี่
dice_rolls = [3, 6, 2]

found = False

for number in dice_rolls:
    if number == 6:
        print("ยินดีด้วย ได้เลข 6!")
        found = True
        break
    else:
        print("ทอยได้", number)
if not found:
     print("หมดโอกาสทอยแล้ว")



````
